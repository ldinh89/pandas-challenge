```python
import pandas as pd
import numpy as np
import os 
cwd = os.getcwd()
print(cwd)

file_to_load = "purchase_data.csv"

purchase_data_df = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
unique_name =purchase_data_df["SN"].unique() 
total_player = len(unique_name)
Sum_player = pd.DataFrame({"Total Players" : [total_player]})
Sum_player
```

```python
unique_item = len(purchase_data_df["Item ID"].unique())
average_price = purchase_data_df["Price"].mean()
number_purchase = purchase_data_df["Purchase ID"].count()
total_revenue = purchase_data_df["Price"].sum()
```

```python
pd.options.display.float_format = '${:,.2f}'.format
summary_dicts =[
{"Number of Unique Items" : unique_item,
 "Average Price" : round(average_price,2),
 "Number of Purchases" : number_purchase,
 "Total Revenue" : total_revenue}]
summary_df = pd.DataFrame(summary_dicts)
summary_df
```

```python
pd.options.display.float_format = '{:,.2f}%'.format
gender_grouped = purchase_data_df.groupby("Gender")
unique_genders = gender_grouped["SN"].nunique()
percentage_players = (unique_genders/total_player) * 100
summary_gender_df = pd.DataFrame(
                              {"Total Count" : unique_genders,
                                  "Percentage of players" : percentage_players
                               })  
                                
summary_gender_df
```

```python
pd.options.display.float_format = '${:,.2f}'.format
purchase_count = gender_grouped["SN"].count()
average_purchase_price = gender_grouped["Price"].mean()
total_purchase_count = gender_grouped["Price"].sum()
average_total_person = total_purchase_count / unique_genders

summary_gender_purchase_df = pd.DataFrame(
                                    {"Purchase Count": purchase_count,
                                     "Average Purchase Price" : average_purchase_price,
                                     "Total Purchase Value" : total_purchase_count,
                                     "Avg Total Purchase per Person" : average_total_person})


summary_gender_purchase_df
```

```python
age_bins = [0,9.90,14.90,19.90,24.90,29.90,34.90,39.90,999]
age_limit = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data_df["Age Demographics"] = pd.cut(purchase_data_df["Age"], age_bins, labels = age_limit, include_lowest = True)
purchase_data_df.head()

pd.options.display.float_format = '{:,.2f}%'.format
age_grouped = purchase_data_df.groupby("Age Demographics")
unique_age = age_grouped["SN"].nunique()
percentage_age = (unique_age / total_player) *100


summary_age_df= pd.DataFrame({"Total Count" : unique_age,
                            "Percentage of Players" : percentage_age})
summary_age_df
```

```python
pd.options.display.float_format = '${:,.2f}'.format
age_purchase_count = age_grouped["SN"].count()
age_average_purchase_price = age_grouped["Price"].mean()
age_total_purchase_count = age_grouped["Price"].sum()
age_average_total_person = age_total_purchase_count / unique_age

summary_age_purchase_df = pd.DataFrame(
                                    {"Purchase Count": age_purchase_count,
                                     "Average Purchase Price" : age_average_purchase_price,
                                     "Total Purchase Value" : age_total_purchase_count,
                                     "Avg Total Purchase per Person" : age_average_total_person})


summary_age_purchase_df
```

```python
pd.options.display.float_format = '${:,.2f}'.format
name_grouped = purchase_data_df.groupby("SN")
name_count = name_grouped["Age"].count()
name_avg_purchase = name_grouped["Price"].mean()
name_total_purchase = name_grouped["Price"].sum()

top_spender_df= pd.DataFrame({"Purchase Count" : name_count,
                              "Average Purchase Price" : name_avg_purchase,
                              "Total Purchase Value" : name_total_purchase})
top_spender_df = top_spender_df.sort_values("Total Purchase Value", ascending = False)
top_spender_df.head()
```

```python
pd.options.display.float_format = '${:,.2f}'.format
popular_item_df = purchase_data_df.loc[:,["Item ID", "Item Name", "Price"]]
item_grouped = popular_item_df.groupby(["Item ID","Item Name"])
item_count = item_grouped["Item ID"].count()
item_avg_purchase = item_grouped["Price"].mean()
item_total_purchase = item_grouped["Price"].sum()

most_popular_item_df = pd.DataFrame({"Purchase Count" : item_count,
                              "Average Purchase Price" : item_avg_purchase,
                              "Total Purchase Value" : item_total_purchase})
most_popular_item_df = most_popular_item_df.sort_values("Purchase Count", ascending = False)

most_popular_item_df.head()
```
